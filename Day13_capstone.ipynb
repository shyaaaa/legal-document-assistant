Legal Document Assistant

Introduction
This project is an AI-powered Legal Document Assistant designed to help paralegals and junior lawyers quickly extract information from legal documents. 

Instead of manually reading large volumes of text, users can ask questions and get accurate answers based only on the provided documents.

The system uses a Retrieval-Augmented Generation (RAG) approach to ensure that responses are grounded in the knowledge base and do not hallucinate information.

Problem Statement

Domain: Legal Document Assistant  
User: Paralegal / Junior Lawyer  

Legal professionals spend significant time reviewing lengthy legal documents to find relevant information. This process is time-consuming and inefficient.

The goal is to build an AI assistant that:
- Answers questions from legal documents
- Avoids hallucination
- Provides accurate and context-based responses
- Handles follow-up queries using memory

```python
DOCUMENTS = [
    {"id": "doc1", "text": "A contract is a legally binding agreement requiring offer, acceptance, and consideration."},
    {"id": "doc2", "text": "Contracts may be written or oral, but written contracts are easier to enforce."},
    {"id": "doc3", "text": "Breach of contract occurs when one party fails to fulfill obligations."},
    {"id": "doc4", "text": "Legal remedies include damages, injunctions, and specific performance."},
    {"id": "doc5", "text": "Tort law deals with civil wrongs like negligence and defamation."},
    {"id": "doc6", "text": "Negligence is failure to exercise reasonable care resulting in damage."},
    {"id": "doc7", "text": "Intellectual property includes patents, copyrights, and trademarks."},
    {"id": "doc8", "text": "Criminal law deals with offenses against the state such as theft or assault."},
    {"id": "doc9", "text": "Legal procedures include filing cases, presenting evidence, and hearings."},
    {"id": "doc10", "text": "Evidence law governs admissibility and reliability of proof in court."}
]

print("Knowledge Base Loaded Successfully")
```

```python
def retrieve_docs(query):
    query = query.lower()
    results = []

    for doc in DOCUMENTS:
        if any(word in doc["text"].lower() for word in query.split()):
            results.append(doc["text"])

    return results[:3]
```

```python
def ask(question):
    context = retrieve_docs(question)

    if not context:
        return "I don't know based on the provided documents."

    return "Answer based on documents:\n" + "\n".join(context)
```

```python
questions = [
    "What is a contract?",
    "What is negligence?",
    "Explain legal remedies",
    "What is intellectual property?",
    "Who is PM of India?"  # Out-of-scope
]

for q in questions:
    print("\nQ:", q)
    print("A:", ask(q))
```

Observations

- The system correctly retrieves relevant information from the knowledge base.
- It does not generate answers outside the provided documents.
- For out-of-scope queries, it responds with "I don't know".
- The system is fast and efficient due to lightweight retrieval.

The Legal Document Assistant successfully demonstrates how AI can be used to simplify legal document analysis.

It reduces the effort required to read large documents and provides accurate answers using a retrieval-based approach.
